```python
Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
MongoDB over SQL databases?
Answer--ongoDB is a popular open-source NoSQL database management system that is designed for storing, retrieving,
and managing large volumes of unstructured or semi-structured data. 
Non-relational databases are a type of database management system 
that do not rely on the traditional tabular relational database model used in SQL databases.

Scenarios where MongoDB is preferred over SQL databases:

Flexible Schema
Horizontal Scalability
Big Data and Real-time Analytics
Agile Development
Geospatial Data 
Content Management Systems (CMS)
Internet of Things (IoT)
Caching

Q2. State and Explain the features of MongoDB.
Answer--MongoDB is a popular NoSQL database management system known for its flexibility, scalability, and ease of use.
Here are some of its key features:

A)Schema-less Design:
MongoDB is schema-less, allowing you to store data without a fixed structure. Documents within a collection can have different
fields and data types,making it flexible for evolving data requirements.
B)JSON-like Documents:
Data in MongoDB is stored in BSON (Binary JSON) format, which is a binary representation of JSON documents. 
This makes it easy to work with data in a format that's similar to JavaScript objects.
Document-Oriented:
MongoDB uses a document-oriented data model, where data is stored in collections of JSON-like documents.
Each document can contain nested arrays and subdocuments, 
making it suitable for complex and hierarchical data structures.

C)Dynamic Schema:
You can add or remove fields from documents without affecting other documents in the same collection.
This flexibility is particularly useful in agile development environments.

D)Scalability:
MongoDB is designed for horizontal scalability. You can distribute data across multiple servers or clusters to
handle large datasets and high traffic loads. 
This is achieved through features like sharding.

E)Automatic Failover:
MongoDB supports automatic failover through replica sets. In the event of a primary server failure, 
a secondary server can automatically take over to ensure high availability and data redundancy.

F)Rich Query Language:
MongoDB provides a powerful query language that supports a wide range of queries, including geospatial queries,
text searches, and aggregation pipelines for complex data manipulation.
G)Indexes:
MongoDB allows you to create indexes on fields to improve query performance. It supports various types of indexes,
including compound indexes and geospatial indexes.

H)Geospatial Capabilities:
MongoDB has built-in support for geospatial indexing and queries, making it suitable for location-based applications and services.

I)Aggregation Framework:
MongoDB's aggregation framework allows you to perform complex data transformations, aggregations, 
and computations within the database itself, reducing the need for client-side processing.


J)Full-Text Search:
MongoDB offers full-text search capabilities, enabling you to perform text searches on text-rich data fields.

K)Replication:
MongoDB supports data replication for redundancy and high availability. It uses replica sets to maintain multiple copies of data across servers.

L)Security:
MongoDB provides robust security features, including authentication, role-based access control, SSL/TLS encryption, and auditing to protect your data.

M)Community and Enterprise Editions:
MongoDB offers both a free and open-source Community Edition and a paid Enterprise Edition with additional features, support, and management tools.
Drivers and Ecosystem:

Q3. Write a code to connect MongoDB to Python. Also, create a database and a collection in MongoDB.
Answer--import pymongo

# Replace these with your MongoDB connection details
mongo_uri = "mongodb://localhost:27017/"  # MongoDB URI
db_name = "mydatabase"  # Name of the database
collection_name = "mycollection"  # Name of the collection

# Connect to MongoDB
try:
    client = pymongo.MongoClient(mongo_uri)
    print("Connected to MongoDB successfully!")
except pymongo.errors.ConnectionFailure as e:
    print("Could not connect to MongoDB: %s" % e)

# Create or access the database
db = client[db_name]

# Create or access the collection
collection = db[collection_name]

# Insert a document into the collection
document = {"name": "John", "age": 30, "city": "New York"}
insert_result = collection.insert_one(document)

# Check if the document was inserted successfully
if insert_result.inserted_id:
    print(f"Document inserted with ID: {insert_result.inserted_id}")
else:
    print("Failed to insert document")

# Close the MongoDB connection
client.close()


Q4. Using the database and the collection created in question number 3, write a code to insert one record,
and insert many records. Use the find() and find_one() methods to print the inserted record.
Answer--import pymongo

# Replace these with your MongoDB connection details
mongo_uri = "mongodb://localhost:27017/"  # MongoDB URI
db_name = "mydatabase"  # Name of the database
collection_name = "mycollection"  # Name of the collection

# Connect to MongoDB
client = pymongo.MongoClient(mongo_uri)

# Access the database and collection
db = client[db_name]
collection = db[collection_name]

# Insert one record
record_one = {"name": "Alice", "age": 25, "city": "San Francisco"}
insert_result_one = collection.insert_one(record_one)

# Check if the record was inserted successfully
if insert_result_one.inserted_id:
    print(f"Inserted record with ID: {insert_result_one.inserted_id}")

# Insert multiple records
records_many = [
    {"name": "Bob", "age": 28, "city": "Los Angeles"},
    {"name": "Carol", "age": 22, "city": "Chicago"},
    {"name": "David", "age": 30, "city": "Houston"},
]

insert_result_many = collection.insert_many(records_many)

# Check if the records were inserted successfully
if insert_result_many.inserted_ids:
    print(f"Inserted {len(insert_result_many.inserted_ids)} records")

# Find and print one record
found_record_one = collection.find_one({"name": "Alice"})
print("Found record one:")
print(found_record_one)

# Find and print all records
found_records = collection.find()
print("Found records:")
for record in found_records:
    print(record)

# Close the MongoDB connection
client.close()

Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to
demonstrate this.
Answer--The find() method in MongoDB is used to query a collection and retrieve documents that match a specified set of criteria.
This method returns a cursor, which can be iterated to access the documents that meet the query conditions
import pymongo

# Replace these with your MongoDB connection details
mongo_uri = "mongodb://localhost:27017/"  # MongoDB URI
db_name = "mydatabase"  # Name of the database
collection_name = "mycollection"  # Name of the collection

# Connect to MongoDB
client = pymongo.MongoClient(mongo_uri)

# Access the database and collection
db = client[db_name]
collection = db[collection_name]

# Find documents that match specific criteria
# In this example, we'll find all documents with "age" greater than or equal to 25
query = {"age": {"$gte": 25}}

# Use the find() method to execute the query
cursor = collection.find(query)

# Iterate through the cursor to access the matching documents
print("Matching documents:")
for document in cursor:
    print(document)

# Close the MongoDB connection
client.close()


Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.
Answer--n MongoDB, the sort() method is used to specify the order in which documents should be returned when querying a collection. 
It allows you to sort the retrieved documents based on one or more fields in ascending or descending order.

query: Specifies the query criteria to filter documents.
projection (optional): Specifies which fields to include or exclude in the results. You can use projection to limit the fields returned, but it's not directly related to sorting.
sort_key: Specifies the field(s) by which the documents should be sorted.
sort_direction: Specifies the sorting order, which can be either pymongo.ASCENDING 


import pymongo

# Replace these with your MongoDB connection details
mongo_uri = "mongodb://localhost:27017/"  # MongoDB URI
db_name = "mydatabase"  # Name of the database
collection_name = "students"  # Name of the collection

# Connect to MongoDB
client = pymongo.MongoClient(mongo_uri)

# Access the database and collection
db = client[db_name]
collection = db[collection_name]

# Define the sort criteria
sort_key = "age"  # Sort by the "age" field
sort_direction = pymongo.DESCENDING  # Sort in descending order

# Use the find() method with sort() to retrieve sorted documents
sorted_students = collection.find().sort(sort_key, sort_direction)

# Iterate through the sorted documents and print them
print("Sorted student records (descending order by age):")
for student in sorted_students:
    print(student)

# Close the MongoDB connection
client.close()


Q7. Explain why delete_one(), delete_many(), and drop() is used.
Answer--In MongoDB, the methods delete_one(), delete_many(), and drop() are used to remove documents or collections from a database.
delete_one(filter):

A)delete_one()
is used to delete a single document from a collection that matches the specified filter or criteria.
It removes the first document that matches the filter and then stops. If multiple documents match the filter, only one will be deleted.
Common use cases for delete_one() include removing individual records or documents based on specific criteria.

collection.delete_one({"name": "John"})

B)delete_many(filter):

delete_many() is used to delete multiple documents from a collection that match the specified filter or criteria.
It removes all documents that match the filter.
Common use cases for delete_many() include batch deletions or clearing out a set of documents that meet certain conditions.

collection.delete_many({"status": "inactive"})

C)
```
